```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import metpy

# Assuming 'xx' is your original xarray dataset

# Step 1: Define the Lambert Conformal projection with the correct parameters
projection = ccrs.LambertConformal(
    central_longitude=262.5,
    central_latitude=38.5,
    standard_parallels=(38.5, 38.5),
    globe=ccrs.Globe(semimajor_axis=6371229, semiminor_axis=6371229)
)

# Step 2: Assign the CRS to the dataset
xx = xx.metpy.assign_crs(
    grid_mapping_name="lambert_conformal_conic",
    latitude_of_projection_origin=38.5,
    longitude_of_central_meridian=262.5,
    standard_parallel=(38.5, 38.5),
    earth_radius=6371229
)

# Step 3: Assign y and x coordinates, forcing the overwrite
xx = xx.metpy.assign_y_x(force=True)

# Check the coordinates
print("Y Coordinates:", xx.y)
print("X Coordinates:", xx.x)

# Step 4: Create the plot using the original projection
fig, ax = plt.subplots(subplot_kw={'projection': projection})

# Plot the data, specifying the correct x and y coordinates
img = xx['dswrf'].plot(ax=ax, transform=projection, cmap='viridis', add_colorbar=True)

# Set the title and labels
ax.set_title('DSWRF Plot in Lambert Conformal')
ax.set_xlabel('X Coordinate (meters)')
ax.set_ylabel('Y Coordinate (meters)')

# Add gridlines
ax.gridlines(draw_labels=True)

# Show the plot
plt.show()

# Step 5: To reproject to WGS 84 (if necessary)
wgs84_projection = ccrs.PlateCarree()

# Create new latitude and longitude grids for WGS 84
new_lon = np.linspace(-180, 180, 100)
new_lat = np.linspace(-90, 90, 100)

# Interpolate the data onto the new grid (this may require a more targeted grid based on your area)
new_grid = xr.Dataset({
    'longitude': (['lon'], new_lon),
    'latitude': (['lat'], new_lat),
})

# Interpolate the dswrf data
dswrf_wgs84 = xx['dswrf'].interp(
    x=new_grid['longitude'], 
    y=new_grid['latitude'],
    method='linear'
)

# Step 6: Plot the reprojected data
fig, ax = plt.subplots(subplot_kw={'projection': wgs84_projection})

# Plot the reprojected data
img = dswrf_wgs84.plot(ax=ax, transform=wgs84_projection, cmap='viridis', add_colorbar=True)

# Set the title and labels
ax.set_title('Reprojected DSWRF to WGS 84')
ax.set_xlabel('Longitude (degrees)')
ax.set_ylabel('Latitude (degrees)')

# Add gridlines
ax.gridlines(draw_labels=True)

# Show the plot
plt.show()
```
